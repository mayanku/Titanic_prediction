```python
print("Before", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)
train_df_filter=train_df.drop(['Ticket','Cabin'],axis=1)
test_df_filter=test_df.drop(['Ticket','Cabin'],axis=1)
combine_filter=[train_df_filter,test_df_filter]
print("After",train_df_filter.shape,test_df_filter.shape,combine_filter[0].shape,combine_filter[1].shape)
```

```python
for dataset in combine_filter:
    dataset['Title']=dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
    
pd.crosstab(train_df_filter['Title'],train_df_filter['Sex'])    
```

```python
for dataset in combine_filter:
    dataset['Title']=dataset['Title'].replace(['Capt','Col','Lady','Countess','Don','Dr','Jonkheer','Lady','Major','Rev','Sir'],'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

train_df_filter[['Title','Survived']].groupby(['Title'],as_index=False).mean()   
```

```python
title_mapping={"Mr":1,"Miss":2,"Mrs":3,"Master":4,"Rare":5}
for dataset in combine_filter:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

train_df_filter.head()
```

```python
train_df_filter = train_df_filter.drop(['Name', 'PassengerId'], axis=1)
test_df_filter = test_df_filter.drop(['Name'], axis=1)
combine_filter = [train_df_filter, test_df_filter]
train_df_filter.shape, test_df_filter.shape
```

```python
for dataset in combine_filter:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)

train_df_filter.head()
```

```python
grid = sns.FacetGrid(train_df_filter, row='Pclass', col='Sex', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for dataset in combine_filter:
    for i in range(0, 2):
        for j in range(0, 3):
            guess_df = dataset[(dataset['Sex'] == i) & \
                                  (dataset['Pclass'] == j+1)]['Age'].dropna()

            # age_mean = guess_df.mean()
            # age_std = guess_df.std()
            # age_guess = rnd.uniform(age_mean - age_std, age_mean + age_std)

            age_guess = guess_df.median()

            # Convert random age float to nearest .5 age
            guess_ages[i,j] = int( age_guess/0.5 + 0.5 ) * 0.5
            
    for i in range(0, 2):
        for j in range(0, 3):
            dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),\
                    'Age'] = guess_ages[i,j]

    dataset['Age'] = dataset['Age'].astype(int)

train_df_filter.head()
```

```python
train_df_filter['AgeBand'] = pd.cut(train_df_filter['Age'],5)
train_df_filter[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
for dataset in combine_filter:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train_df_filter.head()
```

```python
train_df_filter = train_df_filter.drop(['AgeBand'], axis=1)
combine_filter = [train_df_filter, test_df_filter]
train_df_filter.head()
```

```python
for dataset in combine_filter:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1

train_df_filter[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine_filter:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1

train_df_filter[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index=False).mean()
```

```python
train_df_filter = train_df_filter.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
test_df_filter = test_df_filter.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
combine_filter = [train_df_filter, test_df_filter]

train_df_filter.head()
```

```python
for dataset in combine_filter:
    dataset['Age*Class'] = dataset.Age * dataset.Pclass

train_df_filter.loc[:, ['Age*Class', 'Age', 'Pclass']].tail(10)
```

```python
freq_port = train_df_filter.Embarked.dropna().mode()[0]
freq_port
for dataset in combine_filter:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)
```

```python
train_df_filter[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine_filter:
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

train_df_filter.head()
```

```python
test_df_filter['Fare'].fillna(test_df_filter['Fare'].dropna().median(), inplace=True)
test_df_filter.head()
```

```python
train_df_filter['FareBand'] = pd.qcut(train_df_filter['Fare'], 4)
train_df_filter[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
for dataset in combine_filter:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train_df_filter = train_df_filter.drop(['FareBand'], axis=1)
combine_filter = [train_df_filter, test_df_filter]
    
train_df_filter.head(10)
```

```python
X_train = train_df_filter.drop("Survived", axis=1)
Y_train = train_df_filter["Survived"]
X_test  = test_df_filter.drop("PassengerId", axis=1).copy()
X_train.shape, Y_train.shape, X_test.shape
```

```python
# Logistic Regression
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
#checking coefficients in logistic regression
coeff_df=pd.DataFrame(train_df_filter.columns.delete(0))
coeff_df.columns=['Feature']
coeff_df['Correlation']=pd.Series(logreg.coef_[0])
coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
# support vector machine
svc=SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```
